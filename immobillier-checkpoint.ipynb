# questions

Nombre d’appartements et Maisons vendus en 2022

```python
sql_query = ''' SELECT COUNT(*) AS nombre, Type_local
FROM ventes
LEFT JOIN bien ON bien.ID_Local = ventes.ID_Local
WHERE Type_local LIKE 'Maison%' or Type_local LIKE 'Appartement%' AND Type_vente LIKE 'Vente%'
GROUP BY Type_local '''

df = pd.read_sql(sql_query, conn)

df
```

2/ Nombre de biens vendus par trimestre

```python
sql_query = ''' SELECT QUARTER(STR_TO_DATE(Date, '%d/%m/%Y')) AS Trimestre,
       COUNT(*) AS Nombre_de_biens_vendus
FROM ventes
GROUP BY Trimestre
ORDER BY Trimestre; '''

df = pd.read_sql(sql_query, conn)
df
```

4/ Proportion d’appartements vendus par nombre de pièces

```python
sql_query = ''' SELECT
COUNT(*) / (SELECT COUNT(*) FROM ventes WHERE ID_Local LIKE 'A%')* 100 AS proportions, Nombre_pieces_principales
from bien
WHERE Type_local LIKE 'Appartement%'
GROUP BY Nombre_pieces_principales '''

df = pd.read_sql(sql_query, conn)
df
```

5/ Les 10 départements où il y a eu le plus de ventes immobilières

```python
sql_query = ''' Select Département, count(*) nb_ventes
from localisation
INNER JOIN ventes ON localisation.ID_Local = ventes.ID_Local
WHERE Type_vente LIKE 'Vente%'
GROUP BY Département
ORDER BY nb_ventes DESC
LIMIT 10 '''

df = pd.read_sql(sql_query, conn)
df
```

6/ Les 10 départements où il y en a eu le moins

```python
sql_query = ''' Select Département, count(*) nb_ventes
from localisation
INNER JOIN ventes ON localisation.ID_Local = ventes.ID_Local
WHERE Type_vente LIKE 'Vente%'
GROUP BY Département
ORDER BY nb_ventes ASC
LIMIT 10 '''

df = pd.read_sql(sql_query, conn)
df
```
